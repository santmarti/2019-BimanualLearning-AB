```python
from mynotebooks import *
%matplotlib inline

exp = Box2DWorld.ExpSetup(pos_nao=(0,0),pos_obj=(0,1.5),obj_type="box",salientMode="minimum", name='human') # minimum
learn = RobotLearning.LearningModel(exp)
nao,obj = exp.nao, exp.obj
learn.load()
SM = learn.SM
exp.start()
learn.myWorldLoop(5)
```

```python
s = SM.getSalient(1)
errlist = s.errorManager.load_errors(0)
errfwd = [e[0] for e in errlist]
errinv = [e[1] for e in errlist]
n = len(errfwd)

colors=['#0022ff','#008800']
labels=[r'{\large Forward prediction}',
        r'{\large Inverse prediction}']

#fig,ax=VectorFigUtils.makeFigure([0,10,0,1])
xlim=[0,n]
VectorFigUtils.drawPlotY(errfwd,xlim=xlim,colors=colors,labels=labels)
#VectorFigUtils.drawPlotY(errinv,xlim=xlim)
gd = plt.legend(fontsize=23,loc='upper right',bbox_to_anchor=(1.63, 1.055))
plt.axis([xlim[0],xlim[1],0,5])
plt.show()
```

```python
errlist =  learn.errorManager.load_eval(2)
error_mean = [e[0] for e in errlist]
error_var = [e[1] for e in errlist]
n = len(error_mean)

plt.close()
plt.axes(xlim=[0,n],ylim=[0.5,2])
VectorFigUtils.drawPlotXY(range(len(error_mean)),error_mean,error_var,xlabel="trials",ylabel="mean error ($p_i,v_i$)")
plt.show()
```
