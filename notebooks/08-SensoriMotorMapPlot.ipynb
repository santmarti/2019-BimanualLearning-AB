```python
from mynotebooks import *
%matplotlib inline

exp = Box2DWorld.ExpSetup(pos_nao=(0,0),pos_obj=(0,4),obj_type="box",name="human",salientMode='minimum') 
nao = exp.nao
learn = RobotLearning.LearningModel(exp)
learn.load()

w,h=640,480
m = [np.flipud(learn.SM.getStateMat(i).transpose()) for i in range(8)]
nao.restPosition(online=False)
exp.setObjPos([0, 4],0.1)
learn.myWorldLoop(20)
```

```python
fig, ax = Box2DWorld.makeFigure(axes=[-2.5,2.5,-0.5,2.5])
Box2DWorld.plotAll(ax,centers=exp.getSalient(),specials=[0,3],label="s",mainfont=16)
VectorFigUtils.drawMaps(ax,m[0],m[3],extent=[-5,5,-1.5,6],alpha1=0.4,alpha2=0.4)
plt.axis('off')
VectorFigUtils.savefig(fig,ax,'imgs/map0.png')
```
